```python
##Task 1: Plotting Sine and Cosine Curves######################################################################################

import numpy as np
from matplotlib import pyplot as plt
import pandas as pd  ##for dataframes
import seaborn as sn ##for Heatmaps

# Generate a sequence of 1000 evenly spaced numbers from -2π to 2π
#linespace is very useful, since 
x = np.linspace(-2 * np.pi, 2 * np.pi)

# Calculate the sine of these numbers
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the sine and cosine waves
style=("ggplot")
plt.figure(figsize=(16,9))
plt.plot(x, y1, label='sin(x)', color="blue", linewidth=3)
plt.plot(x, y2, label='cos(x)', color="orange", linewidth=3)
plt.title('Plotting Sine and Cosine Curves', fontsize=26, pad=32)
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc="lower left", ) 
plt.grid(True, axis="y", color="gray" )
plt.show()
```

```python
## 2. Add Titles and Labels: Improve the readability of the plot by adding an appropriate title to the histogram.
##   Also, label the x-axis as 'Value' and the y-axis as 'Frequency'.

## 3. Apply Different Color: Change the color of the histogram bars to something other than the default. You may choose any color
##    that you find visually appealing.

## 4. Include Mean Line: Calculate the mean of the data array and draw a vertical line on the histogram at the mean value. Use a contrasting color
##for the line and include a legend to indicate that this line represents the mean.

import numpy as np

# Generate random numbers from a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot histogram
plt.hist(data,bins=30)
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=1.5)
plt.suptitle('Normal Distribution Data: Histogram', fontsize=16)
plt.xlabel('Value', fontsize=14, color='darkblue', labelpad=15)
plt.ylabel('Frequency', fontsize=14, color='darkgreen', labelpad=15)
plt.show()
```
