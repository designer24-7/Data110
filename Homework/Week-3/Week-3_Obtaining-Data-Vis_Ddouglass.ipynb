```python
##Task 1: Plotting Sine and Cosine Curves######################################################################################

import numpy as np
from matplotlib import pyplot as plt
import pandas as pd  ##for dataframes
import seaborn as sn ##for Heatmaps

# Generate a sequence of 1000 evenly spaced numbers from -2π to 2π
#linespace is very useful, since 
x = np.linspace(-2 * np.pi, 2 * np.pi)

# Calculate the sine of these numbers
y1 = np.sin(x)
y2 = np.cos(x)

# Plot the sine and cosine waves
style=("ggplot")
plt.figure(figsize=(16,9))
plt.plot(x, y1, label='sin(x)', color="blue", linewidth=3)
plt.plot(x, y2, label='cos(x)', color="orange", linewidth=3)
plt.title('Plotting Sine and Cosine Curves for Domains from -2π to 2π', fontsize=26, pad=32)
plt.xlabel('x', color="darkgreen", fontsize=20)
plt.ylabel('y', color="navy", fontsize=20)
plt.legend(loc="lower left", fontsize=18) 
plt.grid(True, axis="y", color="gray" )
plt.show()
```

```python
##TASK 2: Enhancing a Histogram of Normal Distribution Data##################################################################

## 1. Modify Bin Count: Experiment with the bins parameter by choosing at least two different values other than 30. Observe how the histogram's appearance changes 
##    with more or fewer bins.

import numpy as np

# Generate random numbers from a normal distribution
data = np.random.normal(0, 1, 1000)

# Plot histogram
plt.hist(data,bins=35, color='orange')
plt.title('Histogram: Random Normal Values \n Between 1 and 1000', fontsize=16, pad=32)
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=1.5)
plt.xlabel('Value', color="darkgreen", fontsize=16, labelpad=15)
plt.ylabel('Frequency', color="navy", fontsize=16, labelpad=15)
plt.legend(['Mean Line']) 
plt.show()
```
